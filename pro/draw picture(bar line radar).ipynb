```python
import pandas as pd
import pyecharts.options as opts
from pyecharts.charts import Radar
data1 = pd.read_csv("data_ave_GDP.csv")
#print(data1)
data2 = pd.read_csv("data_sumGDP.csv")
data3 = pd.read_csv("data_rate.csv")
v0 = [[data1.iloc[0]["2021"],data2.iloc[0]["2021"],data3.iloc[0]["2021"]]]
v1 = [[data1.iloc[1]["2021"],data2.iloc[1]["2021"],data3.iloc[1]["2021"]]]
v2 = [[data1.iloc[2]["2021"],data2.iloc[2]["2021"],data3.iloc[2]["2021"]]]
v3 = [[data1.iloc[3]["2021"],data2.iloc[3]["2021"],data3.iloc[3]["2021"]]]
print(v0,v1,v2,v3)
Radar1 = (
    Radar()
    .add_schema(
        schema=[
            opts.RadarIndicatorItem(name="aveGDP",max_=50000),
            opts.RadarIndicatorItem(name="sumGDP",max_=2000000000000),
            opts.RadarIndicatorItem(name="rate",max_=30),
        ]
    )
    .add('MJL',v0,color="red")
    
    .add('SA',v1,color="#f9713c")
    .add('GER',v2,color="#4169E1")
    .add('QAT',v3,color="00BFFF")
    .set_global_opts(title_opts=opts.TitleOpts(title='radar'),)


)
Radar1.render_notebook()
```

```python
data4 = pd.read_csv("data_agri.csv")
y0 = [0 for i in range(20)]
y1 = [0 for i in range(20)]
for i in range(20):
    y0[i] = data1.iloc[4][i+4]

    y1[i] = data4.iloc[4][i+44]
import pyecharts.options as opts
from pyecharts.charts import Line
from pyecharts.charts import Bar
x = [i+2000 for i in range(20)]
linee=(
    Line()
    .add_xaxis(xaxis_data=x)
    .add_yaxis(series_name='aveGDP increase',y_axis=1000*y0)
    .add_yaxis(series_name='agri increase',y_axis=1000*y1,symbol='arrow',is_symbol_show=True)
    .set_global_opts(title_opts=opts.TitleOpts(title='relationship between agri and aveGDP'))

)
#print(x,y0,y1)

linee.render_notebook()

bar=(
    Bar()
    .add_xaxis(x)
    .add_yaxis("aveGDP",y0,color='blue')
    .add_yaxis("agri",y1*10)
    .set_global_opts(title_opts=opts.TitleOpts(title="relationship"))

)
bar.render_notebook()
```
